### Empresas e microempresas
Enquanto empresas não possuem CNPJ e não CPF, microempresários possuem um CPF e não CNPJ, com isso, sabendo do número total de fornecedores no SICAF, podemos calcular quantas microempresas são cadastradas em relação ao total

```python
com_cpf = df['CPF'].count()
com_cnpj = df['CNPJ'].count()
print(f"Há um total de {com_cpf + com_cnpj} empresas cadastradas")
print(f"Dessas, {com_cpf} são micro-empresas")
print(f"e {com_cnpj} não são")
print(f"{com_cpf/50} a cada {com_cnpj/50} empresas cadastradas são micro-empresas")
```
